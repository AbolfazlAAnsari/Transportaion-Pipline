```python
# Set input values 

simulation.input['Temperature(F)'] = 10
simulation.input['Humidity(%)'] = 99

simulation.input['Wind_Chill(F)'] = 1
simulation.input['Pressure(in)'] = 0
simulation.input['Visibility(mi)'] = 0
simulation.input['Wind_Speed(mph)'] = 0
simulation.input['Precipitation(in)'] = 0
```
