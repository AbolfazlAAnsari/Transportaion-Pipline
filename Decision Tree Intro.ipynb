#### Decision Trees (DTs):
They are a non-parametric supervised learning method used for <br>
classification and regression. <br>
The objective is to create a classifier to predict the <br>
label of future data point based on the rules that trained.<br>

The deeper the tree, the more complex the decision rules and the fitter the model.<br> It's the risk of overfitting in high depth trees.

#### Advantages

Simple to understand and interpret <Br>
Requires little data preparation. ( however does not support missing values ) <Br>
The computation cost is logarithmic, because of the structure of tree <Br>
Able to handle both numerical and categorical data.<br> ( however sklearn does not support categorical ones )<Br>



#### Disadvantages

They can create over-complex trees that do not generalize the data well. <Br>
Small variation in tree can cause complexity, the solution is using an ensemble. <Br>
The problem of learning an optimial decision tree is NP-complete. In more practical examples of<br> decision tree, you should use heuristics algorithms such as greedy ones. <br>So, there is not guarante of returning Optimal Solution.
Some concepts are hard to learn such as XOR.<br>
    
Decision tree learners create biased trees if some classes dominte.<Br>It is better of balanced the dataset prior to fitting with the decision tree.

#### Decision Tree Classifier 
DecisionTreeClassfier is a class capable of peroformaing multi-class classification on a dataset.<br>

#### Syntax
to fit a decision tree classfier, you should provide two arrays: <br>
first one X array, holding the training samples with the lables ( n_samples, n_features ) <br>
and the second one is Y array with shape ( n_samples, ) 

```python
from sklearn import tree
X = [[0,0], [1,1]]
Y = [0, 1]
dtree = tree.DecisionTreeClassifier()
dtree = dtree.fit(X, Y)
```

After being fitted, the model can be used to predict the class samples:

```python
dtree.predict([[2,2]])
```

When there are multiple class with highest and same probability for <br>
prediction, the classifier return the class with lowest index.

Capable of doing both binary and multi-class classification.

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, Y = iris.data, iris.target
# Iris is a dataset with four features and three labels with 150 records

dtree = tree.DecisionTreeClassifier()
dtree = dtree.fit(X, Y)
```

```python
iris
```

Once trained, you can plot the tree with the plot_tree function:

```python
tree.plot_tree(dtree)
```

#### DecisionTreeClassifiers Parameters

criterion: {'gini', 'entropy', 'log_loss'}, default='gini'<br>
splitter: {'best', 'random'}, default='best'<br>
max_depth: int, default=None
min_samples_split: int or float, default=2<br>
max_leaf_nodes: int, default=None<br>

min_impurity_decrease: float, deafult=0.0<br>
a node will be split if this split induces a decrease of the impurity greater<br> than or equal to this value.<br>
The weighted impurity decrease equation: <br>
N_t / N * (impurity - N_t_R / N_t * right_impurity
                    - N_t_L / N_t * left_impurity)<br>
